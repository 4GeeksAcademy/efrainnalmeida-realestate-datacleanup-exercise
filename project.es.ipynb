#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
# Normalizar los precios por población usando Min-Max Scaling
ds_subconjunto["price_norm"] = ds_subconjunto.groupby("level5")["price"].transform(
    lambda x: (x - x.min()) / (x.max() - x.min())
)

# Crear el histograma con 4 poblaciones en un solo gráfico
plt.figure(figsize=(8, 5))

# Trazar histogramas para cada población
ds_subconjunto.groupby("level5")["price_norm"].plot(kind="hist", alpha=0.5, bins=10, legend=True)

# Agregar etiquetas y título
plt.title("Histogramas Normalizados de Precios por Población")
plt.xlabel("Precio Normalizado")
plt.ylabel("Frecuencia")
plt.legend()

# Mostrar el gráfico
plt.show()
```

```python
from ipyleaflet import Map, Marker

# Filtrar solo las poblaciones del cinturón sur de Madrid
ds_subconjunto = ds[ds["level5"].isin(["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"])].copy()

# Reemplazar comas por puntos y convertir a float en las columnas de coordenadas
ds_subconjunto["latitude"] = ds_subconjunto["latitude"].str.replace(",", ".").astype(float)
ds_subconjunto["longitude"] = ds_subconjunto["longitude"].str.replace(",", ".").astype(float)

# Crear un diccionario con las coordenadas de cada población
coordenadas_dict = ds_subconjunto.groupby("level5")[["latitude", "longitude"]].apply(lambda x: x.values.tolist()).to_dict()

from ipyleaflet import CircleMarker

# Diccionario de colores para cada población
colores = {"Fuenlabrada": "red", "Leganés": "blue", "Getafe": "green", "Alcorcón": "purple"}

# Agregar los puntos al mapa usando apply(), con colores distintos por población
ds_subconjunto.apply(lambda row: map.add_layer(CircleMarker(
    location=(row["latitude"], row["longitude"]),
    radius=6,  # Tamaño del marcador
    color=colores.get(row["level5"], "black"),  # Color según población
    fill=True,
    fill_color=colores.get(row["level5"], "black"),
    fill_opacity=0.7,
    title=row["level5"]
)), axis=1)

# Mostrar el mapa con los puntos agregados
map
```
